Importing the Dependencies 

```python
import numpy as np
import pandas as pd 
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Preprocessing

```python
#loading the data
movies_data = pd.read_csv('movies.csv')
```

```python
#selecting relevant features 

features = ['genres', 'keywords', 'tagline', 'cast', 'director']
features
```

```python
#replacing null values with null strings


for feature in features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combine all the selected features 

combined_features = movies_data['genres'] + ' ' + movies_data['keywords'] + ' ' + movies_data['tagline'] + ' ' + movies_data['cast'] + ' ' + movies_data['director']
```

```python
combined_features
```

```python
#converting texture data to feature vector 

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity 

```python
#Getting the similiarity score

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
#getting the movie name from the user 

movie_name = input('Enter your favorite movie name: ')
```

```python
#creating a list with all the movie names given in the dataset

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match for the movie name given by the user


close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(close_match)
```

```python
close_match_fin = close_match[0]
print(close_match_fin)
```

```python
#finding the index of the movie with title

index_of_movie = movies_data[movies_data.title == close_match_fin]['index'].values[0]
print(index_of_movie)
```

```python
#getting a list of similar movies 

similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on similarity score

sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse= True)
print(sorted_similar_movies)
```

```python
#using the name of similar movies based on index

print('Movies Suggested for you : \n')

i = 1 

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if i < 20 :
        print(i, '.',title_from_index)
        i += 1
```

Movie Recommendation System

```python
movie_name = input('Enter your favorite movie name: ').title()

list_of_all_titles = movies_data['title'].tolist()

close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match_fin = close_match[0]

index_of_movie = movies_data[movies_data.title == close_match_fin]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_movie]))

sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse= True)


print('Movies Suggested for you : \n')

i = 1 

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if i < 20 :
        print(i, '.',title_from_index)
        i += 1
```
